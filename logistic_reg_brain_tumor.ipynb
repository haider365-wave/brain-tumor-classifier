```python
sns.countplot(x = "Class" , data=df ,palette="mako" , hue="Class")
plt.title('Class distrpution 0 for no tumor 1 for tumor')
```

```python
X = df.drop("Class" , axis=1)
y = df["Class"]
X.describe()
```

```python
fig, axes = plt.subplots(4, 4, figsize=(18, 15))
axes = axes.flatten()
for i, col in enumerate(X.columns):
    sns.histplot(X[col], kde=True, ax=axes[i] , color="DarkGreen")
    axes[i].set_title(col)
plt.suptitle('Feature Distributions')
plt.tight_layout()
plt.show()
```
